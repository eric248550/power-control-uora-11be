```python
import matplotlib.pyplot as plt
import numpy as np
import random

# New simulation parameters
x_values = range(1, 151)  # Varying x from 1 to 20
C = 3  # Capture threshold constant remains the same
gamma = 2  # Path loss exponent remains the same
iterations_per_x = 1000  # Reduced iterations for quicker computation
r=15

Lmax = 5
p_min = 10
p_max = 250

# Modified simulation function to accept variable x
def simulate_for_x(x):
    results = []
    for _ in range(iterations_per_x):
#         W = np.random.uniform(1, 10, x)
#         W = 1
        W = p_min + ((p_max-p_min)/(Lmax-1)) * random.choice(range(1, Lmax + 1, 1))
#         D = 1 + np.random.uniform(0, r, x)
        D = np.random.uniform(0, r, x)
        power_ratios = W * (D**(-gamma))
        
        max_power = np.max(power_ratios)
        sum_others = np.sum(power_ratios) - max_power
        
        ratio = max_power / sum_others
        results.append(ratio > 10**(C/10))
    print(results)
    return np.mean(results)

# Run simulations for varying x
probability_estimates = [simulate_for_x(x) for x in x_values]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_values, probability_estimates, marker='o', linestyle='-', color='blue')
plt.title('Capture Probability vs. Number of Signals')
plt.xlabel('Number of Signals (x)')
plt.ylabel('Estimated Capture Probability ($P_{cap}$)')
plt.grid(True)
plt.xticks(x_values)
plt.show()
```
